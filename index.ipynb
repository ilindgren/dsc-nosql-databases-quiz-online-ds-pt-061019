## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MongoDB is an open source database management system (DBMS) that uses a document-oriented database model which supports various forms of data. ... Instead of using tables and rows as in relational databases, the MongoDB architecture is made up of collections and documents. A record in MongoDB is a document, which is a data structure composed of field and value pairs. The fields in documents are akin to the columns in a relational database, and the values they contain can be a variety of data types, including other documents, arrays and arrays of documents, according to the MongoDB user manual.

Documents, which also must incorporate a primary key as a unique identifier, are the basic unit of data in MongoDB. Collections contain sets of documents and function as the equivalent of relational database tables. Collections can contain any type of data, but the restriction is the data in a collection cannot be spread across different databases.



## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault-tolerant technology is a capability of a computer system, electronic system or network to deliver uninterrupted service, despite one or more of its components failing. Fault tolerance also resolves potential service interruptions related to software or logic errors. The purpose is to prevent catastrophic failure that could result from a single point of failure.

A Resilient Distributed Dataset is a collection of an arbitrary number of server clusters that share a full dataset distributed between them. With multiple servers comes multiple opportunities for things like server failures, and our dataset is no good if we have no guarantee that part of our dataset might be missing when we go to query for it. RDDs have multiple servers with backups of each server on other servers so we can query all servers simultaniously without failures.
